# **1. Import Dataset** #


```python
import pandas as pd


df = pd.read_csv('heart_disease_uci.csv')
```

# **2. Eksplorasi Awal** 
	

```python
# Tampilkan 5 baris pertama
print(df.head())

# Tampilkan 5 baris terakhir
print(df.tail())

# Informasi tentang dataset
print(df.info())

# Cek nilai missing
print(df.isnull().sum())

# Deteksi outliers menggunakan IQR
import numpy as np

# Pilih hanya kolom numerik
df_numeric = df.select_dtypes(include=[np.number])

# Deteksi outliers menggunakan IQR pada kolom numerik
Q1 = df_numeric.quantile(0.25)
Q3 = df_numeric.quantile(0.75)
IQR = Q3 - Q1

# Tentukan outliers berdasarkan IQR
outliers = df_numeric[(df_numeric < (Q1 - 1.5 * IQR)) | (df_numeric > (Q3 + 1.5 * IQR))]
print(outliers)

```

# **3. Statistik Deskriptif** 


```python
import pandas as pd


df = pd.read_csv('heart_disease_uci.csv')


# Statistik deskriptif
mean = df.mean(numeric_only=True)
median = df.median(numeric_only=True)
mode = df.mode(numeric_only=True).iloc[0]
std_dev = df.std(numeric_only=True)
variance = df.var(numeric_only=True)
skewness = df.skew(numeric_only=True)
kurtosis = df.kurt(numeric_only=True)

print(f"Mean:\n{mean}\n")
print(f"Median:\n{median}\n")
print(f"Mode:\n{mode}\n")
print(f"Standar Deviasi:\n{std_dev}\n")
print(f"Variansi:\n{variance}\n")
print(f"Skewness:\n{skewness}\n")
print(f"Kurtosis:\n{kurtosis}\n")

# Korelasi antar variabel numerik
df_numeric = df.select_dtypes(include=[float, int])
correlation = df_numeric.corr()
print(f"Korelasi:\n{correlation}\n")
```

# **4. Visualisasi Data** 

```python
import pandas as pd
import matplotlib.pyplot as plt

# Muat dataset
df = pd.read_csv('heart_disease_uci.csv')

# Pilih hanya kolom numerik
df_numeric = df.select_dtypes(include=[float, int])

# Hitung jumlah kolom numerik
num_cols = len(df_numeric.columns)

# Tentukan layout berdasarkan jumlah kolom
rows = (num_cols // 3) + 1
cols = min(num_cols, 3)

# Plot box plot untuk setiap variabel numerik
df_numeric.plot(kind='box', subplots=True, layout=(rows, cols), figsize=(15, 10))
plt.tight_layout()  # Untuk menghindari overlapping antar subplot
plt.show()
```
